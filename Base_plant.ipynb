```python
def reform_table(data):
    data.reset_index().drop('index',axis=1)
    data['parts'].unique().tolist()
    df = pd.DataFrame(columns=data['parts'].unique().tolist())
    
    for i in range(int(data.shape[0]/9)):
        df.loc[i,'w_of_y'] = data['w_of_y'].iloc[i*9+0]
        df.loc[i,'plant_id'] = data['plant_id'].iloc[i*9+0]
        df.loc[i,'초장'] = data['values'].iloc[i*9+0]
        df.loc[i,'엽장'] = data['values'].iloc[i*9+1]
        df.loc[i,'엽폭'] = data['values'].iloc[i*9+2]
        df.loc[i,'엽병장'] = data['values'].iloc[i*9+3]
        df.loc[i,'엽수'] = data['values'].iloc[i*9+4]
        df.loc[i,'관부직경'] = data['values'].iloc[i*9+5]
        df.loc[i,'화방 꽃수(소화수)'] = data['values'].iloc[i*9+6]
        df.loc[i,'착과수'] = data['values'].iloc[i*9+7]
        df.loc[i,'최종화방차수'] = data['values'].iloc[i*9+8]

    data = data.drop(['parts','unit','values'],axis=1)
    data = data.drop_duplicates().reset_index().drop('index',axis=1)
    
    data = data.merge(df,on=['w_of_y','plant_id'],how='left')
    data['d_of_w'] = data['checkdate'].apply(lambda x:pd.Period(x).day_of_week)
    return data    
```

```python
t_plant_data = plant_data[plant_data['farm_id']=='B농가']
t_plant_data
```

```python
data = reform_table(t_plant_data)
data.to_pickle('./p_d/B_plant.pkl')
```
